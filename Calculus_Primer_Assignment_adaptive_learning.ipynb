```python
# using bisection search
import numpy as np
import matplotlib.pyplot as plt
import time

def f(x):
    return x**2 - 4*x + 3

def gradient(x):
    return 2*x - 4

# Set the convergence threshold for interval size
convergence_threshold = 1e-9

cost=[]

def binary_search(low,high):

  start_time= time.time()

  while (high-low) > convergence_threshold:

    mid= low+(high-low)/2
    cost.append(f(mid))

    if gradient(mid)>0:
      high=mid
    else:
      low=mid

  end_time= time.time()
  elapsed_time = end_time-start_time
  return (low,high,elapsed_time)

optimized_x,optimized_y,elapsed_time = binary_search(-5,5)
print(optimized_x)
print(optimized_y)
print((optimized_x+optimized_y)/2)
print(f((optimized_x+optimized_y)/2))
print(elapsed_time)

plt.plot(cost)
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.show()
```
